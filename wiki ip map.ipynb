```python
def getLinks(articlelink):
    page = urlopen('http://wikipedia.org' + articlelink)
    s = BeautifulSoup(page)
    return s.find('div',{'id':'bodyContent'}).findAll('a',href = re.compile('^(/wiki/)((?!:).)*$'))
def getHistoryIPs(pageURL):
    pageURL = pageURL.replace('/wiki/','')
    history = "http://en.wikipedia.org/w/index.php?title="+pageURL+"&action=history"
    print('history link: '+history)
    page = urlopen(history)
    s = BeautifulSoup(page)
    ips = set()
    for ip in s.findAll('a',{'class':'mw-anonuserlink'}):
        ips.add(ip.text)
    return ips
links = getLinks('/wiki/Kevin_Bacon')
while len(links)>0:
    for link in links:
        print ('－－－叫我分割线－－－')
        historyIPs = getHistoryIPs(link.attrs['href'])
        for ip in historyIPs:
            print(ip)
    newLink = links[random.randint(0,len(links)-1)].attrs['href']
    links = getLinks(newLink)
```

```python
def getCountry(ip):
    try:
        response = urlopen('http://freegeoip.net/json/'+ip).read().decode('utf-8')
    except urllib.error.HTTPError:
        return None
    country = json.loads(response)
    return country.get('country_code')
links = getLinks('/wiki/Ewha_Womans_University')
while len(links)>0:
    for link in links:
        print('----我是分割线----')
        ips = getHistoryIPs(link.attrs['href'])
        for ip in ips:
            country = getCountry(ip)
            if country is not None:
                print ('ip '+ip+' is from '+ country)
    newLink = links[random.randint(0,len(links)-1)].attrs['href']
    links = getLinks(newLink)
```

```python
import os
from urllib.request import urlretrieve
```

```python
downloadDirectory = 'downloaded'
baseUrl = 'http://pythonscraping.com'
def getAbsoluteURL(baseUrl, source):
    if source.startswith('http://www.'):
        url = source[11:]
    elif source.startswith('http://'):
        url = source
    elif source.startswith('www.'):
        url = source[4:]
        url = 'http://'+source
    else:
        url = baseUrl + '/' + source
    if baseUrl not in url:
        return None
    return url
def getDownloadPath(baseUrl,absoluteUrl,downloadDirectory):
    path = absoluteUrl.replace('www.','')
    path = path.replace(baseUrl,'')
    path = downloadDirectory + path
    directory = os.path.dirname(path)
    
    if not os.path.exists(directory):
        os.makedirs(directory)
        
    return path

html = urlopen('http://www.pythonscraping.com')
bsobj = BeautifulSoup(html)
downloadList = bsobj.findAll(src = True)

for download in downloadList:
    fileUrl = getAbsoluteURL(baseUrl,download['src'])
    if fileUrl is not None:
        print(fileUrl)

urlretrieve(fileUrl,getDownloadPath(baseUrl,fileUrl,downloadDirectory))
```
