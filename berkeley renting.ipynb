```python
def room(row):
    bd = row.split('·')[0]
    ba = row.split('·')[1]
    sqft = row.split('·')[2]
    bd = int(bd.split(' ')[0])
    ba = float(ba.split(' ')[1])
    if '--' in sqft:
        sqft = np.nan
    elif int(sqft.split(' ')[1])<100:
        sqft = np.nan
    else:
        sqft = float(sqft.split(' ')[1])
    return pd.Series({'bedroom':bd,'bathroom':ba,'area_in_sqft':sqft})
rooms = su_rent['Photocard value 3'].apply(room)
```

```python
def postcode(row):
    code = str(row).split('.')[0]
    return pd.Series({'Postcode':code})
```

```python
postcode = su_rent['Hide number'].apply(postcode)
```

```python
def price(row):
    rent = row.split('/')[0]
    rent = rent[1:]
    rent = rent.split(',')
    rent = ''.join(rent)
    rent= int(rent)
    return pd.Series({'rent':rent})
```

```python
price = su_rent['Photocard value 2'].apply(price)
```
