```python
import string
import re
import operator
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
def isCommon(ngram):
    commonWords = ['body','mr',"the", "be", "and", "of", "a", "in", "to", "have", "it","i", "that", "for", "you", "he", "with", "on", "do", "say", "this","they", "is", "an", "at", "but","we", "his", "from", "that", "not","by", "she", "or", "as", "what", "go", "their","can", "who", "get","if", "would", "her", "all", "my", "make", "about", "know", "will","as", "up", "one", "time", "has", "been", "there", "year", "so","think", "when", "which", "them", "some", "me", "people", "take","out", "into", "just", "see", "him", "your", "come", "could", "now","than", "like", "other", "how", "then", "its", "our", "two", "more","these", "want", "way", "look", "first", "also", "new", "because","day", "more", "use", "no", "man", "find", "here", "thing", "give","many", "well"]
    for word in ngram:
        if word in commonWords:
            return True
    return False
```

```python
def CleanInput(data):
    data = re.sub('\n+','',data).lower()
    data = re.sub('\[[0-9]*\]','', data)
    data= re.sub(' +',' ', data)
    data = bytes(data, 'UTF-8')
    data = data.decode('ascii','ignore')
    cleaned_data = []
    data = data.split(' ')
    for d in data:
        d = d.strip(string.punctuation)
        if len(d)>0 or (d.lower() == 'a' or d.lower() == 'i'):
            cleaned_data.append(d)
    return cleaned_data
```

```python
def ngram(data, n):
    data= CleanInput(data)
    output = {}
    for i in range(0,len(data)-n+1):
        if not isCommon(data[i:i+n]):
            ngram_data = ' '.join(data[i:i+n])
            if ngram_data not in output:
                output[ngram_data] = 0
            output[ngram_data]+=1
    return output
```

```python
from random import randint
def wordlistsum(wordlist):
    total = 0
    for word, value in wordlist.items():
        total+=value
    return total
```

```python
def randomword(wordlist):
    rand = randint(1,wordlistsum(wordlist))
    for word,value in wordlist.items():
        rand -=value
        if rand<=0:
            return word
    
```

```python
def worddictbuilder(text):
    text = text.replace('\n',' ')
    text = text.replace('\\',' ')
    punctuations = [',','.',';',':']
    for s in punctuations:
        text = text.replace(s,' '+s+' ')
    word = text.split(' ')
    word = [w for w in word if w!='']
    worddict = {}
    for i in range(1,len(word)):
        if word[i-1] not in worddict:
            worddict[word[i-1]] = {}
        if word[i] not in worddict[word[i-1]]:
            worddict[word[i-1]][word[i]] = 0
        worddict[word[i-1]][word[i]] +=1
    return worddict
```

```python
text = str(urlopen("http://pythonscraping.com/files/inaugurationSpeech.txt").read(),'utf-8')
worddict = worddictbuilder(text)
```

```python
length = 100
current = 'I'
chain = ''
for i in range(0,length):
    chain += current +' '
    current = randomword(worddict[current])
```
